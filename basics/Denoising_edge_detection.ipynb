```python
import cv2 
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('noisy_staff.jpg', 1)
```

```python
plt.imshow(img)
```

```python
kernel = np.ones((6,6),np.float32)/25
filt_2D = cv2.filter2D(img,-1,kernel)    #Convolution using the kernel we provide
plt.imshow(filt_2D)
```

```python
kernel = np.ones((3,3),np.float32)/25
filt_2D = cv2.filter2D(img,-1,kernel)    #Convolution using the kernel we provide
plt.imshow(filt_2D)
```

```python
kernel = np.ones((3,3),np.float32)/25
filt_2D = cv2.filter2D(img,-1,kernel)    #Convolution using the kernel we provide
plt.imshow(filt_2D)
```

```python
blur = cv2.blur(img,(3,3))
plt.imshow(blur)
```

```python
blur_gaussian = cv2.GaussianBlur(img,(3,3),0)
plt.imshow(blur_gaussian)
```

```python
median_blur = median = cv2.medianBlur(img,3)
plt.imshow(median_blur)
```

```python
bilateral_blur = cv2.bilateralFilter(img,9,50,75)
plt.imshow(median_blur)
```

```python
plt.imshow(img)
plt.imshow(filt_2D)
plt.imshow(blur)
plt.imshow(blur_gaussian)
plt.imshow(median_blur)
plt.imshow(bilateral_blur)
```
