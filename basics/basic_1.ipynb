#Saving Numpy array into a text file

```python

arr_reshaped = img.reshape(img.shape[0], -1)
np.savetxt("img.txt", arr_reshaped)
```

#Loading a image from text

```python
loaded_img = np.loadtxt("img.txt")
loaded_img = loaded_img.reshape( loaded_img.shape[0], loaded_img.shape[1] // 3, img.shape[2])
```
